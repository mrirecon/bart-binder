# BART example using bash shell

```bash
bart version
```

Create oversampled radial trajectory.

```bash
bart traj -r -y55 -x256 trajectory
```

```bash
bart scale 0.5 trajectory trajectory2
```

Simulate k-space data.

```bash
bart phantom -t trajectory2 -k kspace
```

Fourier transform a long radial spoke and show sinogram.

```bash
bart fft -i 2 kspace sinogram
```

```bash
bart toimg sinogram sinogram.png
```

```bash
cat sinogram.png | display
```

Compute Ram-Lak filter and apply to kspace.

```bash
bart rss 1 trajectory2 ramlak
```

```bash
bart fmac kspace ramlak kspace_dcf
```

Apply adjount nufft (gridding) and show result.

```bash
bart nufft -a trajectory2 kspace_dcf img
```

```bash
bart toimg img img.png
```

```bash
cat img.png | display
```
